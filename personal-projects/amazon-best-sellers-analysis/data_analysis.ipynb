```python
# contando quantas ocorr√™ncias de cada g√™nero dos livros mais bem avaliados
max_rating = books.ratings.max()
best_books = books.query(f"ratings=={max_rating}")
best_books.genre.value_counts()
```

```python
# contando quantas ocorr√™ncias de cada g√™nero de todos os livros
books.genre.value_counts()
```

```python
# best sellers de 2021
best_sellers_2021 = books.query("year==2021")
best_sellers_2021
```

### üìà Parte 2: Visualiza√ß√£o de dados

```python
# desenha a figura e determina seu tamanho
fig = plt.figure(figsize=(15, 10))
plt.show()
```

```python
# lista com os anos (2009 a 2021)
years = list(books.year.unique())

# lista vazia que vai armazenar o n√∫mero 
# de avalia√ß√µes por ano
ratings_count = []

# loop que calcula n√∫mero de avalia√ß√µes por ano
for item in years:
    # cria um dataframe com o ano que est√° sendo percorrido
    year_dataframe = books.query(f"year=={item}")
    # calcula a soma das avalia√ß√µes de cada livro daquele ano
    count = year_dataframe.reviews.sum()
    # adiciona a soma na lista
    ratings_count.append(count)

print(ratings_count)
```

```python
first_chart = pd.DataFrame(ratings_count, years)
first_chart.reset_index(inplace=True)
first_chart.rename(columns={'index': 'year', 0: 'ratings_count'}, inplace=True)
```

```python
first_chart
```

```python
fig = plt.figure(figsize=(15, 5))
plt.plot(years, ratings_count)

plt.title('N√∫mero total de avalia√ß√µes por ano entre os 100 best sellers de cada ano')

plt.ylabel('N√∫mero de avalia√ß√µes')
plt.xlabel('Ano')
```

```python
# primeiro gr√°fico
fig = plt.figure(figsize=(15, 5))
plt.title('Top 100 Best Sellers de 2009 a 2021',pad=15,size=25, color='grey')
plt.ylabel('n√∫mero de avalia√ß√µes')
plt.xlabel('posi√ß√£o no ranking')
plt.bar(books['ranks'], books['reviews'], width=0.5, alpha=0.7, color='#042c8b')
plt.show()
```
