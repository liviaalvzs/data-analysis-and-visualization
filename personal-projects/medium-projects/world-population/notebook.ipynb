## 游깴 **An치lise e Visualiza칞칚o de Dados Sobre a Popula칞칚o Mundial entre 1951 - 2020**

### Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importando e tratando o dado

```python
main_df = pd.read_csv("src/world-population.csv")
main_df
```

```python
# convertendo os valores de "yearly_growth_percentage" para float
replace_pattern = r'\$|%'  
main_df['yearly_growth_percentage'].replace(replace_pattern, '', regex=True, inplace=True) 
main_df['yearly_growth_percentage'] = main_df['yearly_growth_percentage'].astype(float)
```

### 游늳 Plotando os Gr치ficos

##### 1. Plotando Gr치ficos Simples

**a) Gr치fico relacionando o ano e a popula칞칚o mundial** 

```python
fig = plt.figure(figsize=(15, 5))
plt.plot(main_df.year, main_df.population.values)

plt.title('World Population Over the Years (1951 - 2020)')

plt.ylabel('Population (Billion)')
plt.xlabel('Year')
```

**b) Gr치fico relacionando o ano e a popula칞칚o urbana**

```python
fig = plt.figure(figsize=(15, 5))
plt.plot(main_df.year, main_df.urban_population.values)

plt.title('World Urban Population Over the Years (1951 - 2020)')

plt.ylabel('Population (Billion)')
plt.xlabel('Year')
```

**c) melhorando a visualiza칞칚o dos gr치ficos criados acima (itens a e b)** 

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(15, 5))
plt.plot(main_df.year, main_df.population.values, '#264653', linewidth=2.8)

plt.title('World Population Over the Years (1951 - 2020)', pad=20, size=16)

plt.ylabel('Total Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(15, 5))
plt.plot(main_df.year, main_df.urban_population.values, '#734f5a', linewidth=2.8)

plt.title('World Urban Population Over the Years (1951 - 2020)', pad=20, size=16)

plt.ylabel('Urban Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.show()
```

##### 2. Plotando m칰ltipos dados em um mesmo gr치fico

**a) Gr치fico relacionando a popula칞칚o urbana e a popula칞칚o total ao passar dos anos**

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)

plt.plot(main_df.year, main_df.population.values, '#264653', linewidth=2.8, label='Total Population')
plt.plot(main_df.year, main_df.urban_population.values, '#734f5a', linewidth=2.8, label='Urban Population')

plt.title('World Total and Urban Population Over the Years (1951 - 2020)', pad=20, size=16)

plt.ylabel('Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.legend(fontsize=12)

plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.show()
```

**b) Gr치fico relacionando a porcentagem da popula칞칚o urbana e a densidade populacional ao passar dos anos**

```python
plt.style.use('seaborn-v0_8-darkgrid')

fig, a =  plt.subplots(figsize=(15,5))

a.bar(main_df.year, main_df.yearly_growth_percentage, alpha=0.7, label='Pop. Growth (%)', color='#2a9d8f', width=0.69)
a.yaxis.set_major_locator(plt.MaxNLocator(10))
a.set_ylabel('Pop. Growth (%)', labelpad=10, size=13)
a.set_xlabel('Year', labelpad=5, size=13)

a1 = a.twinx()

a1.plot(main_df.year, main_df.pop_density_km2, '#264653', linewidth=2.8, label='Pop. Density (km)', alpha=0.7)
a1.grid(False) 
a1.set_ylabel('Pop. Density (km)', labelpad=10, size=13)

plt.title('Yearly Population Density and Growth (1951 - 2020)', pad=20, size=16)

fig.legend(loc="upper right", bbox_to_anchor=(0.075,1), bbox_transform=ax.transAxes)

plt.show()
```
