## 游깴 **An치lise e Visualiza칞칚o de Dados Sobre a Popula칞칚o Mundial entre 1951 - 2020**

### Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importando e tratando o dado

```python
main_df = pd.read_csv("src/world-population.csv")
main_df.head()
```

```python
# convertendo os valores de "yearly_growth_percentage" para float
main_df['yearly_growth_percentage'].replace(r'%', '', regex=True, inplace=True) 
main_df['yearly_growth_percentage'] = main_df['yearly_growth_percentage'].astype(float)
```

### 游늳 Plotando os Gr치ficos Separadamente

##### 1. Plotando Gr치ficos Simples

**a) Gr치fico relacionando o ano e a popula칞칚o mundial** 

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(main_df.year, main_df.population)

plt.title('World Population Over the Years (1951 - 2020)')

plt.ylabel('Population (Billion)')
plt.xlabel('Year')

plt.tight_layout()
plt.savefig('teste2', dpi=200)
```

**b) Gr치fico relacionando o ano e a popula칞칚o urbana**

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(main_df.year, main_df.urban_population.values)

plt.title('World Urban Population Over the Years (1951 - 2020)')

plt.ylabel('Population (Billion)')
plt.xlabel('Year')

plt.tight_layout()
plt.savefig('teste3', dpi=200)
```

**c) melhorando a visualiza칞칚o dos gr치ficos criados acima (itens a e b)** 

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(10, 5))
plt.plot(main_df.year, main_df.population, color='#9BA17B', linewidth=2.8)

plt.title('World Population Over the Years (1951 - 2020)', pad=20, size=16)
plt.ylabel('Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

plt.tight_layout()
plt.savefig('teste5', dpi=200)
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(10, 5))
plt.plot(main_df.year, main_df.urban_population, color='#9BA17B', linewidth=2.8)

plt.title('World Urban Population Over the Years (1951 - 2020)', pad=20, size=16)
plt.ylabel('Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

plt.tight_layout()
plt.savefig('teste4', dpi=200)
```

##### 2. Plotando m칰ltipos dados em um mesmo gr치fico

**a) Gr치fico relacionando a popula칞칚o urbana e a popula칞칚o total ao passar dos anos**

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(10, 5))

plt.plot(main_df.year, main_df.population.values, '#9BA17B', linewidth=2.8, label='Total Population')
plt.plot(main_df.year, main_df.urban_population.values, '#797d62', linewidth=2.8, label='Urban Population')

plt.title('World Total and Urban Population Over the Years (1951 - 2020)', pad=20, size=16)

plt.ylabel('Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.legend(fontsize=12)

plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(10, 5))

plt.plot(main_df.year, main_df.population.values, '#9BA17B', linewidth=2.8, label='Total Population')
plt.plot(main_df.year, main_df.urban_population.values, '#797d62', linewidth=2.8, label='Urban Population')

plt.title('World Total and Urban Population Over the Years (1951 - 2020)', pad=20, size=16)

plt.ylabel('Population (Billion)', labelpad=10, size=13)
plt.xlabel('Year', labelpad=5, size=13)

plt.legend(fontsize=12)

plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.tight_layout()
plt.savefig('teste6', dpi=200)
```

**b) Gr치fico relacionando a porcentagem de crescimento populacional e a densidade populacional ao passar dos anos**

```python
plt.style.use('seaborn-v0_8-darkgrid')

fig, a =  plt.subplots(figsize=(10,5))

a.bar(main_df.year, main_df.yearly_growth_percentage, alpha=0.7, label='Pop. Growth (%)', color='#9BA17B', width=0.69)
a.set_ylabel('Pop. Growth (%)', labelpad=10, size=13)
a.set_xlabel('Year', labelpad=5, size=13)

a1 = a.twinx()

a1.plot(main_df.year, main_df.pop_density_km2, color='#797d62', linewidth=2.8, label='Pop. Density (km)')
a1.grid(False) 
a1.set_ylabel('Pop. Density (km)', labelpad=10, size=13)

plt.title('Yearly Population Density and Growth (1951 - 2020)', pad=20, size=16)

fig.legend(loc="upper right", bbox_to_anchor=(0.30,0.89))
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig, a =  plt.subplots(figsize=(10,5))

plt.title('Yearly Population Density and Growth (1951 - 2020)', pad=20, size=16)

a.bar(main_df.year, main_df.yearly_growth_percentage, alpha=0.7, label='Pop. Growth (%)', color='#9BA17B', width=0.69)
a.set_ylabel('Pop. Growth (%)', labelpad=10, size=13)
a.set_xlabel('Year', labelpad=5, size=13)

a1 = a.twinx()

a1.plot(main_df.year, main_df.pop_density_km2, color='#797d62', linewidth=2.8, label='Pop. Density (km)')
a1.grid(False) 
a1.set_ylabel('Pop. Density (km)', labelpad=10, size=13)


fig.legend(loc="upper right", bbox_to_anchor=(0.26,0.89))
plt.tight_layout()
plt.savefig('teste6', dpi=200)
```

### 游늳 Plotando os gr치ficos em um 칰nico plot

```python
plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots(3, 2, figsize=(20,15))
plt.suptitle('World Population Over The Years (1950-2020)', fontsize=20, y=0.95)
plt.subplots_adjust(hspace=0.35)

# figura 1: 
ax[0,0].set_title('Total Population', pad=15, size=16)
ax[0,0].plot(main_df.year, main_df.population.values, color='#285430', linewidth=2.8)
ax[0,0].set_ylabel('Population (Billion)', labelpad=10, size=13)
ax[0,0].set_xlabel('Year', labelpad=5, size=13)

# figura 2: 
ax[0,1].set_title('Population Density (Km)', pad=15, size=16)
ax[0,1].bar(main_df.year, main_df.pop_density_km2.values, color='#9BA17B', linewidth=2.8)
ax[0,1].set_ylabel('Pop. Density (Km)', labelpad=10, size=13)
ax[0,1].set_xlabel('Year', labelpad=5, size=13)

# figura 3: 
ax[1,0].set_title('Urban Population Compared to Total Population ', pad=15, size=16)
ax[1,0].plot(main_df.year, main_df.population.values, '#285430', linewidth=2.8, label='Total Population')
ax[1,0].bar(main_df.year, main_df.urban_population.values, linewidth=2.8, color='#797d62')
ax[1,0].set_ylabel('Population (Billion)', labelpad=10, size=13)
ax[1,0].set_xlabel('Year', labelpad=5, size=13)

# figura 4: 
ax[1,1].set_title('Urban Population', pad=15, size=16)
ax[1,1].plot(main_df.year, main_df.population.values, linewidth=2.8, color='#797d62')
ax[1,1].set_ylabel('Population (Billion)', labelpad=10, size=13)
ax[1,1].set_xlabel('Year', labelpad=5, size=13)

# figura 5: 
ax[2,0].set_title('Yearly Growth (Percentage)', pad=15, size=16)
ax[2,0].plot(main_df.year, main_df.yearly_growth_percentage	.values, color='#557153', linewidth=2.8)
ax[2,0].set_ylabel('Pop. Growth (%)', labelpad=10, size=13)
ax[2,0].set_xlabel('Year', labelpad=5, size=13)

# figura 6: 
ax[2,1].set_title('Yearly Growth (Percentage and Values)', pad=15, size=16)
ax[2,1].bar(main_df.year, main_df.yearly_growth.values, linewidth=2.8, color='#A9AF7E', label='Pop. Growth (Values)')
ax[2,1].set_ylabel('Pop. Growth (Values)', labelpad=10, size=13)
ax[2,1].set_xlabel('Year', labelpad=5, size=13)
ax[2,1].legend(loc='upper left')

ax2 = ax[2,1].twinx()
ax2.plot(main_df.year, main_df.yearly_growth_percentage	.values, color='#557153', linewidth=2.8, label='Pop. Growth (%)')
ax2.set_ylabel('Pop. Growth (%)', labelpad=10, size=13)
ax2.legend(loc='upper right')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('teste7', dpi=200)
```
